## 4. Categorical Exploration `(9 pts)`

Find the number of Pokémon per primary type.

```python
# put your answer here
df['Type 1'].value_counts()
```

Find the number of Pokémon per generation.

```python
# put your answer here
df['Generation'].value_counts()
```

Which type appears the most? Which appears the least?

```python
# put your answer here
type_counts = df['Type 1'].value_counts()
most_common_type = type_counts.index[0]
least_common_type = type_counts.index[-1]
print(f"The most common type is: {most_common_type}")
print(f"The least common type is: {least_common_type}")
```

How many unique primary types (Type 1) exist?

```python
 # put your answer here
df['Type 1'].nunique()
```

How many unique secondary types (Type 2) exist?

```python
 # put your answer here
df['Type 2'].nunique()
```


Which primary types have the most dual-type combinations?

```python
 # put your answer here
dual_types = df[df['Type 2'].notnull()]
type_counts = dual_types['Type 1'].value_counts()
most_common_type = type_counts.index[0]
print(f"The most common primary types with dual type combinations is: {most_common_type}")
```

Which type has the highest mean Attack?

```python
# put your answer here
type_highest_mean_attack = df.groupby('Type 1')['Attack'].mean().idxmax()
print(f"The type with the highest mean Attack is: {type_highest_mean_attack}")
```

Which type has the lowest mean Defense?

```python
# put your answer here
type_lowest_mean_defense = df.groupby('Type 1')['Defense'].mean().idxmin()
print(f"The type with the lowest mean Defense is: {type_lowest_mean_defense}")
```


Which generation has the highest average Speed?

```python
# put your answer here
generation_highest_mean_speed = df.groupby('Generation')['Speed'].mean().idxmax()
print(f"The generation with the highest average Speed is: {generation_highest_mean_speed}")
```

## 5. Groupby & Aggregation `(13 pts)`

Compute the average Attack per primary type.

```python
# put your answer here
average_attack_per_type = df.groupby('Type 1')['Attack'].mean()
average_attack_per_type
```

Compute the maximum HP per generation.


```python
# put your answer here
max_hp_per_generation = df.groupby('Generation')['HP'].max()
max_hp_per_generation
```

Compute the total number of Pokémon per primary type.

```python
# put your answer here
total_pokemon_per_type = df.groupby('Type 1').size()
total_pokemon_per_type
```

For each type, compute:

- mean Attack

- mean Defense

- mean Speed

```python
# put your answer here
stats_per_type = df.groupby('Type 1')[['Attack', 'Defense', 'Speed']].mean()
stats_per_type
```

For each generation, compute:

- count

- mean Total

- number of Legendary Pokémon (hint: use sum on Boolean)

```python
# put your answer here
stats_per_generation = df.groupby('Generation').agg({'Generation': 'count', 'Total': 'mean', 'Legendary': 'sum'})
stats_per_generation
```

Which type combination (Type 1 + Type 2) has the highest average Attack?

```python
# put your answer here
type_combination_highest_attack = df.groupby(['Type 1', 'Type 2'])['Attack'].mean().idxmax()
print(f"The type combination with the highest average Attack is: {type_combination_highest_attack}")
```

Which generation has the highest proportion of Legendary Pokémon?

```python
# put your answer here
legendary_proportion_per_generation = df.groupby('Generation')['Legendary'].mean()
generation_with_highest_legendary_proportion = legendary_proportion_per_generation.idxmax()
print(f"The generation with the highest proportion of Legendary Pokémon is: {generation_with_highest_legendary_proportion}")
```


Which primary type has the largest variance in HP?

```python
# put your answer here
type_highest_variance_hp = df.groupby('Type 1')['HP'].var().idxmax()
print(f"The primary type with the largest variance in HP is: {type_highest_variance_hp}")
```

Which primary type has the highest median Speed?

```python
# put your answer here
type_highest_median_speed = df.groupby('Type 1')['Speed'].median().idxmax()
print(f"The primary type with the highest median Speed is: {type_highest_median_speed}")
```

Group Pokémon by whether they are Legendary or not. Compare:

- mean Total

- mean Attack

- mean Defense

- mean Speed

```python
# put your answer here
stats_per_legendary = df.groupby('Legendary')[['Total', 'Attack', 'Defense', 'Speed']].mean()
stats_per_legendary
```

Show the top 5 strongest types by mean Total.

```python
# put your answer here
strongest_types = df.groupby('Type 1')['Total'].mean().nlargest(5)
strongest_types
```

Rank generations by their average Attack.

```python
# put your answer here
average_attack_per_generation = df.groupby('Generation')['Attack'].mean().rank()
average_attack_per_generation
```

Show the top 10 fastest Pokémon using nlargest(10, "Speed").

```python
# put your answer here
top_10_fastest = df.nlargest(10, 'Speed')
top_10_fastest
```
